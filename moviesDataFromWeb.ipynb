```python
import pandas as pd
import numpy as np
```

Extracting movies data of 2018 from wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
```

```python
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df
```

```python
#We dont have the genres of the movie in wikipedia, so use TMDB website to get the genres of the movies.
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '87e96b7aa5c777dbb6c76e449b4032d4'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/157336?api_key=87e96b7aa5c777dbb6c76e449b4032d4'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.nan
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
#Also we have to get director and actor names from cast and crew
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df['director_name'] = df['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df['actor_1_name'] = df['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.nan
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df['actor_2_name'] = df['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.nan
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df['actor_3_name'] = df['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df
```

```python
df = df.rename(columns={'Title':'movie_title'})
```

```python
df_18 = df.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
df_18['actor_2_name'] = df_18['actor_2_name'].replace(np.nan, 'unknown')
df_18['actor_3_name'] = df_18['actor_3_name'].replace(np.nan, 'unknown')
```

```python
df_18['movie_title'] = df_18['movie_title'].str.lower()
```

```python
df_18['comb'] = df_18['actor_1_name'] + ' ' + df_18['actor_2_name'] + ' '+ df_18['actor_3_name'] + ' '+ df_18['director_name'] +' ' + df_18['genres']
```
